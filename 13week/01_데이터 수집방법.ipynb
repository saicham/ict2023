```python
# 공항 코드
aircode = ['GMP','PUS','CJU','TAE','KWJ','RSU','USN','KUV','WJU','CJJ']

# 가져올 행의 갯수
num  = 100

#가져올 페이지 번호
npage = 1
```

```python
api = "http://openapi.airport.co.kr/service/rest/AirportParkingCongestion/airportParkingCongestionRT?schAirportCode={Air_code}&serviceKey={Key}&numOfRows={Nrow}&pageNo={Pno}"
```

```python
### 공항별 데이터 수집하기

for air_code in aircode:
#     print(air_code)
    url = api.format(Air_code = air_code,
                    Key = apikey,
                    Nrow = num,
                    Pno = npage)
    print(url)
    
    ### url 요청 및 응답받기
    rs_data = requests.get(url)
#     print(rs_data)

    # url의 정보를 텍스트형태로 읽기
    rs_text = rs_data.text
#     print(rs_text)

     # text데이터를르 XML태그형태로 파싱하기
    rs_xml = BeautifulSoup(rs_text,'lxml')
#     print(rs_xml)

    for i in range(0, len(rs_xml.find_all("parkingairportcodename")),1):
        print("공항명 :",rs_xml.find_all('airportkor')[i].text)
        print("주차장명 :",rs_xml.find_all('parkingairportcodename')[i].text)  
        print("기준일자 :",rs_xml.find_all('sysgetdate')[i].text)  
        print("기준시간 :",rs_xml.find_all('sysgettime')[i].text)  
        print("혼잡율 :",rs_xml.find_all('parkingcongestiondegree')[i].text)          
        print("혼잡도 :",rs_xml.find_all('parkingcongestion')[i].text)     
        print("전체주차면수 :",rs_xml.find_all('parkingtotalspace')[i].text)             
        print("사용중인 주차면수 :",rs_xml.find_all('parkingoccupiedspace')[i].text)                     
        print('----------------------------------------------------------')
```

### 3.웹크롤링 데이터 수집하기

```python
from bs4 import BeautifulSoup as bs
import requests
```

####  요청 사이트에 URL요청 및 응답 받아오기

```python
url = "https://search.naver.com/search.naver?query=날씨"
html = requests.get(url)

```

#### 응답결과는 html코드로 받아옴

```python
html.text
```

#### HTML태그로 인식할수 있도록 파싱하기

```python
soup = bs(html.text,'html.parser')
soup
```

#### 원하는 태그 위치 접근하기
#### div 태그중에 clacc 속성의 값이 scroll_box인것을 찾음

```python
rs_html = soup.find("div",{"class":"scroll_box"})
rs_html
```

#### 날씨 정보 추출하기

```python
for em, blind, num in zip(rs_html.findAll("em"),
                         rs_html.findAll("span",{"class":"blind"}),
                         rs_html.findAll("span",{"class":"num"})):
    print(em.text,blind.text,num.text)
```
